```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline 

#to actually show plot once created instead of just creating them so inline is used
```

```python
import numpy as np
x=np.linspace(0,5,11)
y=x**2
```

```python
x
```

```python
y
```

## Basic Matplotlib commands

```python
plt.plot(x,y,'g')
plt.xlabel("X-AXIS")
plt.ylabel("Y-AXIS")
plt.title("GRAPH")
plt.show() #even if show is not used it will still show the graph since inline is used
```

## Creating Multiplots in same canvas

```python
#plt.subplot(nrows,ncols,plot_number)
plt.subplot(1,2,1)
plt.plot(x,y,'r--')
plt.subplot(1,2,2)
plt.plot(y,x,'g*-')
plt.show()
```

```python
#creating empty canvas
fig=plt.figure()
```

```python
axes=fig.add_axes([0.1,0.1,0.8,0.8]) #left bottom width height (range 0 to 1)
axes.plot(x,y,'b')
plt.show()
```

```python
fig=plt.figure()
axes1=fig.add_axes([0.1,0.1,0.8,0.8])
axes2=fig.add_axes([0.2,0.5,0.4,0.3])
axes1.plot(x,y,'b')
axes1.set_xlabel("X")
axes1.set_ylabel("Y")

axes2.plot(y,x,'r')
axes2.set_xlabel("X")
axes2.set_ylabel("Y")
plt.show()
```

```python
#plt.tightlayout-prevents overlapping
```

```python
fig=plt.figure(figsize=(8,4),dpi=100)
plt.show()
```

```python
fig,axes=plt.subplots(figsize=(12,3))
axes.plot(x,y,'pink')
plt.show()
```

## LEGENDS

```python
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
ax.plot(x,x**2,label="x**2")
ax.plot(x,x**3,label="x**3")
ax.legend(loc=0)
plt.show()
```

```python
#loc(1)-upper right 2-upper left 3-lower left 4-lower right 0-optimal location
```
