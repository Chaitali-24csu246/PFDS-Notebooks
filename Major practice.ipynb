### Missing Data

The four types of data are nominal (labels), ordinal (ranked), interval (equal intervals, no true zero), and ratio (equal intervals with a true zero).

* nominal(labels)
* Ordinal(Ranked)
* Interval(Equal intervals, no true zero)
* Ratio (equal intervals with true zero)
* *categorical(nominal and ordinal)

1️⃣ Nominal Data

Names / labels only — no order

Categories with no ranking

Cannot be meaningfully compared

Examples:

Gender (Male, Female)

Blood group (A, B, AB, O)

Company names (Google, Amazon)

👉 Operations allowed: count, mode

2️⃣ Ordinal Data (often confused — important)

Categorical data WITH order, but unequal gaps

Order matters

Difference between values is not measurable

Examples:

Education level (High, Medium, Low)

Ratings (Poor, Average, Good, Excellent)

Class rank (1st, 2nd, 3rd)

👉 Operations allowed: min, max, median

3️⃣ Interval Data

Ordered numeric data with equal intervals, NO true zero

Differences are meaningful

Zero is arbitrary, not absence

Examples:

Temperature in °C or °F

Calendar years

IQ scores

👉 Cannot say “twice as much”

4️⃣ Ratio Data

Interval data WITH a true zero

Zero means absence

Ratios make sense

Examples:

Height, weight

Age

Income

Distance

👉 All arithmetic operations valid

```python
#composition vs inheritance
#nheritance (IS-A relationship)

#Definition:
#Inheritance occurs when a class (child) derives from another class (parent) and inherits its attributes and
class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def show(self):
        print(f"Name: {self.name}, Age: {self.age}")

# Inheritance
class Student(Person):
    def __init__(self, name, age, student_id):
        super().__init__(name, age)
        self.student_id = student_id

    def show(self):
        super().show()
        print(f"Student ID: {self.student_id}")

s = Student("Alice", 20, 101)
s.show()
```

Definition:
Composition occurs when a class contains objects of other classes as members.

Relationship: “Student HAS-A Address”

```python
class Address:
    def __init__(self, city, country):
        self.city = city
        self.country = country

    def show(self):
        print(f"{self.city}, {self.country}")

class Student:
    def __init__(self, name, age, address):
        self.name = name
        self.age = age
        self.address = address  # composition

    def show(self):
        print(f"Name: {self.name}, Age: {self.age}")
        self.address.show()

addr = Address("New York", "USA")
s = Student("Bob", 22, addr)
s.show()
```
