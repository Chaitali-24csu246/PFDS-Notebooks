#### Create a class 'Student' has three data members which are name, age and address. The constructor of the class assigns default values name as "unknown", age as '0' and address as "not available". It has two member functions with the same name 'setInfo()'. The first method has two parameters for name and age and assigns the same whereas the second method takes three parameters which are assigned to name, age and address respectively. Print the name, age and address of 5 students. (3 Marks)

```python
class Student:
    def __init__(self):
        self.name="Unknown"
        self.age = 0
        self.address='Not available'
    def setInfo(self, name, age, address=None):
        self.name = name
        self.age = age
        if address!=None:
            self.address=address
    def display(self):
        print("Name:", self.name," Age:", self.age,"Address: ",self.address)
students=[]
n=['A','B','C','D','E']
a=[10,38,45,22,19]
add=['ABC Enclave', 'XYZ Society', 'Green Park', 'LMNOPQ House', 'H 1234 Sector 098']
#Adding address to every alternate entry only
for i in range(0,5):
    students.append(Student())
    print("Defaualt setting : ")
    students[i].display()
    if i%2:
        students[i].setInfo(n[i],a[i])
    else:
        students[i].setInfo(n[i],a[i],add[i])
    print("After setInfo")
    students[i].display()   
```

#### Write a program for manipulating coordinates in the Polar coordinate Represent points as objects. The class Polar must include data members such as radius and theta, and add( ), sub( ), angle(), are member functions.
#### 1) add():- Add two objects of Polar class

#### 2) sub():- Subtract two objects of Polar class

#### Hint:- Convert polar coordinates to rectangular (Cartesian) coordinates in order to perform operations on them. Then convert the result back to Polar.

#### convert from Polar Coordinates (r,θ) to Cartesian Coordinates (x,y) : x = r × cos( θ ) y = r × sin( θ ). (4 Marks)

```python
import math
class Polar:
    def __init__(this, radius=0, theta=0):
        this.radius=radius
        this.theta=theta
    def angle(this): #cartesian conversion
        x=this.radius * math.cos(math.radians(this.theta))
        y=this.radius * math.sin(math.radians(this.theta))
        return x, y
    def add(this, other):
        x1,y1=this.angle()
        x2,y2=other.angle()
        x=x1+x2
        y=y1+y2
        r=(x**2+y**2)**0.5
        theta=math.degrees(math.atan2(y,x))
        return Polar(r, theta)
    def sub(this, other):
        neg_other=Polar(other.radius, other.theta + 180)#using 180 to invert a-b = a+(-b) so we use add
        return this.add(neg_other)
    def display(this):
        print("Radius = ", this.radius," Theta = ",this.theta)

p1=Polar(2,3)
p2=Polar(10,24)
print("Polar cooridnate 1:")
p1.display()
print("Polar coordinate 2: " )
p2.display()
print("Adding")
p1.add(p2).display()
print("Subtracting")
p1.sub(p2).display()


        
```

#### A subclass called Cylinder is derived from the superclass Circle as shown in the class diagram (Ref. Figure 1). Implements the program to show the subclass Cylinder invokes the superclass constructors) and inherits the variables and methods from the superclass Circle. It overrides the getArea() method in the subclass Cylinder that computes the surface area ((=2π×radius×height + 2×base-area). It means, if getArea() method is called by the a Circle instance, it returns the base area. If getArea() is called by a Cylinder instance, it returns the surface area of the cylinder

```python
class Circle:
    def __init__(self,radius=0):
        self.radius=radius
    def getRadius(self):
        return self.radius
    def setRadius(self,radius):
        self.radius=radius
    def getArea(self):
        return 3.14*self.radius**2
class Cylinder(Circle):
    def __init__(self,radius, height):
        self.height=height
        super().__init__(radius)
    def getHeight(self):
        return self.height
    def setHeight(self,h):
        self.height=h
    def getArea(self):
        return self.getHeight()*3.14*self.getRadius()*2+2*self.getRadius()**2*3.14
c=Circle(2)
print("Radius:", c.getRadius())
print("Area of Circle:", c.getArea())
cy=Cylinder(2, 10)
print("Radius:", cy.getRadius())
print("Height:", cy.getHeight())
print("Area of Cylinder:",cy.getArea())
```
