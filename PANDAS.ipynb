## Multi Index and Index Hierarchy

```python
#index levels
outside=['G1','G1','G1','G2','G2','G2']
inside=[1,2,3,1,2,3]
hier_index=list(zip(outside,inside))
hier_index=pd.MultiIndex.from_tuples(hier_index)
```

```python
hier_index
```

```python
df=pd.DataFrame(randn(6,2),index=hier_index, columns=('A B').split())
```

```python
df
```

```python
df.loc['G1']
```

```python
df.loc['G1'].loc[1]
```

```python
df.index.names
```

```python
df.index.names=['Group','Num']
df
```

```python
df.xs('G1')#xs means croass section method, it returns cross section
```

```python
df.xs(('G1',2))#does not accept list as a key
```

```python
df.xs(1,level='Num')#returns 1st level of g1 g2
```

```python
df.xs(3,level='Num')
```
