### If we have nan value we cannot proceed further for analysis. we need to reprocess
### easisest: delete. but only value is not deleted, row/col also deleted. data loss
### Dealing with missing data: data reprocessing
### change value: to similar value
### but if we have lots of data, this also fails
### mean median mode come in
### mean: problem is outlier, only numeric values
### boxplot to visualize and find outlier
### if outlier exists, go with median not mean, also only numeric values
### mode, categorical data, freq of occurrence
### check percentage of missing data. if >50% then delete it.

### MERGING JOINING CONCATENATION

3 main ways of combining data frames together.

```python
#by default it is row wise, specify axis for column wise
import pandas as pd
```

```python
df1=pd.DataFrame({'A':['A0','A1','A2','A3'],'B':['B0','B1','B2','B3'],'C':['C0','C1','C2','C3'],'D':['D0','D1','D2','D3']},index=[0,1,2,3])
```

```python
df2=pd.DataFrame({'A':['A4','A5','A6','A7'],'B':['B4','B5','B6','B7'],'C':['C4','C5','C6','C7'],'D':['D4','D5','D6','D7']},index=[4,5,6,7])
```

```python
df3=pd.DataFrame({'A':['A8','A9','A10','A11'],'B':['B8','B9','B10','B11'],'C':['C8','C9','C10','C11'],'D':['D8','D9','D10','D11']},index=[8,9,10,11])
```

```python
#for concatenation dimensions should match along ax,s, pass list of df.
pd.concat([df1,df2,df3])
```

```python
pd.concat([df1,df2,df3],axis=1)#matrix created by column addition, index no match, value missing, nan
```

```python
#if index for all had been 1,2,3,4 then it would have worked without nan
```

#### Merging

```python
left=pd.DataFrame({'key':['K0','K1','K2','K3'],'A':['A0','A1','A2','A3'],'B':['B0','B1','B2','B3']})
left
```

```python
right=pd.DataFrame({'key':['K0','K1','K2','K3'],'C':['C0','C1','C2','C3'],'D':['D0','D1','D2','D3']})
right
```

```python
#merge allows you to merge 2 df using similar logic to sql, at least 2 df reqd. type of join specified
```

```python
pd.merge(left,right,how='inner',on='key')
```

```python
pd.merge(left,right,how='outer',on='key')
```

```python
pd.merge(right,left,how='inner',on='key')
```

```python
pd.merge(right,left,how='outer',on='key')
```

```python
left=pd.DataFrame({'key1':['K0','K1','K2','K3'],'key2':['K0','K1','K0','K1'],'A':['A0','A1','A2','A3'],'B':['B0','B1','B2','B3']})
left
```

```python
right=pd.DataFrame({'key1':['K0','K1','K2','K3'],'key2':['K0','K0','K0','K0'],'C':['C0','C1','C2','C3'],'D':['D0','D1','D2','D3']})
right
```

```python
pd.merge(left,right,on=['key1','key2'],how="inner")
```

```python
pd.merge(left,right,on=['key1','key2'],how="outer")
```

```python
pd.merge(left,right,on=['key1','key2'])
```

```python
#where keys dont match, values dont merge
```

```python
pd.merge(left,right,on=['key1','key2'],how="right")
```
