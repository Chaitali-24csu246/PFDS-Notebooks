```python
l= [10, 20, 30, 40, 50, 60, 70]
i= l.pop(4)  #pop returns a value
l.insert(1, i)
l.append(i)
print("Updated list:",l)
```

```python
l = [1, 2, 3, 4, 5, 6, 7, 8, 9]
a = len(l) // 3   
new = []
for i in range(0, len(l), a):
    part = l[i:i+a]       
    part.reverse()       
    new.extend(part)      
print(new)
```

```python
l=[1,2,3,1,7,4,8,4,6,2,2,0,9,7]
same=list()
d=dict()
for i in l:
    if i not in same:
        same.append(i)
        d[i]=1;
    else:
        d[i]+=1
print(d)
```

```python
l1=["a","b","c","d"]
l2=[1,2,3,4]
s=set()
for i in range(0,len(l1)):
    s.add((l1[i],l2[i]))
print(s)
```

```python
set1 = {1, 2, 3, 4, 5}
set2 = {3, 4, 6, 7}
common= set1.intersection(set2)
print("Common elements:", common)
set1.difference_update(common)
print("after removing ", set1)
```

```python
set1 = {1, 2, 3, 4, 5}
set2 = {2, 3, 4}
#checking if s1 is subset or superset of set 2
if set1.issubset(set2):
    set1.clear()
elif set1.issuperset(set2):
    set1.clear()
print("set1:", set1)
print("set2:", set2)
```

```python
fruit = ['apple', 'banana', 'cherry', 'date']
d = {1: 'apple', 2: 'banana', 3: 'grape'}
k= [key for key, value in d.items() if value in fruit]
for i in k:
    del d[i]

print("Updated dictionary:", d)
```

```python
d = {1: 'apple', 2: 'banana', 3: 'apple', 4: 'cherry'}
r= []
for value in d.values():
    if value not in r:
        r.append(value)
print(r)
```
