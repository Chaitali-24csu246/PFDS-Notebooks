```python
import pandas as pd
```

```python
#create dataframe
data = {'Company':['GOOG','GOOG','MSFT','MSFT','FB','FB'],
        'Person': ['Sam','Charlie','Amy','Vanessa','Carl','Sarah'],
        'Sales':[200,120,340,124,243,350]}
```

```python
data
```

```python
df = pd.DataFrame(data)
```

```python
df
```

```python
df.groupby('Company') #.groupby() method to group rows tgtr based off a column name
```

```python
by_comp= df.groupby("Company")
```

```python
by_comp
```

```python
df
```

```python
by_comp.mean(numeric_only = True) #cant just write mean directly because data has objects and we need mean of numeric only
```

```python
df.dtypes
```

```python
by_comp['Sales'].mean() #this is represented in series
```

```python
by_comp[['Sales']].mean() #this is dataframe
```

```python
by_comp.std
```

```python
by_comp[['Sales']].std()
```

```python
by_comp[['Sales','Person']].min() #for minimum value
```

```python
by_comp[['Sales']].max()
```

```python
by_comp[['Sales','Person']].max()
```

```python
df.loc[by_comp['Sales'].idxmax()]
```

```python
by_comp.count()
```

```python
df.count()
```

```python
by_comp.describe()
#gives statistical description of all the numeric columns inside a df
```

```python
df.describe()
```

```python
by_comp.describe().transpose()
```

```python
by_comp.describe().transpose()['GOOG']
```
