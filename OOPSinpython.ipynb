### Defining class

```python
class Fraction(object):
    def __init__(self,numerator,denominator):
        self.__numerator=numerator
        self.__denominator=denominator
        #self.reduce()
    def __repr__(self):
        return str(self.__numerator)+'/'+str(self.__denominator)
    #getters and seeters created below
    def showvalue(self):
        print(self.__numerator, "/",self.__denominator)
    def getNumerator(self):
        "gets value for numerator"
        return self.__numerator
    def getDenominator(self):
        "gets value for denominator"
        return self.__denominator
    def setNumerator(self,value):
        "sets valuemfor numerator"
        self.__numerator=value
    def setDenominator(self, value):
        "sets value for denominator"
        self.__denominator=value
```

```python
#All values are being taken as string, so no mathematical operation can be performed
n=input("Enter numerator")
d=input("enter denominator")
obj1=Fraction(n,d)
obj1.showvalue()
print(obj1.getNumerator())
print(obj1.getDenominator())
```

```python
#demonstrating use of setters
d=input("Enter new denominator")
n=input("enter new numerator")
print("Old fraction")
obj1.showvalue()
obj1.setNumerator(n)
obj1.setDenominator(d)
print("New fraction")
obj1.showvalue()
```
